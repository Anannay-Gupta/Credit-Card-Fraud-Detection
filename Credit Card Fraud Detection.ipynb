```python
#This dataset is highly unbalanced as label 0(normal transactions) 
#is almost more than 99% of data hence we can't put the data in our model like this so we will need to do some preprocessing.
```

```python
legit = cc_data[cc_data.Class==0]
fraud = cc_data[cc_data.Class==1]
```

```python
#Under-Sampling
#Here we will build a sample dataset from the original dataset containing 
#similliar distribution of normal transactions and fraudalent transactions.
#Number of fraudalent transactions -> 492.
```

```python
legit_sample = legit.sample(n=492)#will take random 492 legit values
```

```python
#Now we will concatenate two df's the fraudalent one and the new random legit one(legit sample)
new_data = pd.concat([legit_sample,fraud],axis = 0)
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
#It shows that the distance is still there in the new dataset between both fraud and legit transactions hence good dataset.
#Very important to check whether we got a good sample or bad sample.
```

```python
X = new_data.drop(columns = 'Class',axis = 1)
Y = new_data['Class']
```

```python
print(X)# Now we dont have our class column in the dataset
```

```python
print(Y) # the class column is stored here in Y
```

```python
#now split the data into testing and training data.
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,stratify = Y,random_state = 2)
```

```python
print(X_train.shape,X_test.shape) #80% of data goes in X_train and 20% into X_test
```

```python
#logisticRegressiontraining
model = LogisticRegression(max_iter=1000)
```

```python
#training logisticregression model with training data
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline

scaler = StandardScaler()
model = LogisticRegression(max_iter=1000)
pipeline = make_pipeline(scaler, model)

pipeline.fit(X_train, Y_train)
```

```python
#Model Evaluation 
```

```python
#Based on Accuracy Score (Accuracy on training data)
X_train_pred = pipeline.predict(X_train)  #predicting labels for training data
training_data_accuracy = accuracy_score(X_train_pred,Y_train) #and then comparing them to original values in Y_train
print('Accuracy on training data: ',training_data_accuracy)
```

```python
#we got accuracy score of 95.2%
```

```python
X_test_pred = pipeline.predict(X_test)
test_data_accuracy = accuracy_score(X_test_pred,Y_test)
print('Accuracy on test data: ',test_data_accuracy)
```

```python
#the score is 91.87%
```
